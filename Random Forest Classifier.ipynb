```python
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml
from sklearn.ensemble import RandomForestClassifier
```

```python
np.random.seed(0)
X = np.genfromtxt('data/X_train.txt', delimiter=None)
Y = np.genfromtxt('data/Y_train.txt', delimiter=None)
X,Y = ml.shuffleData(X,Y)
Xtr,Ytr,Xva,Yva = X[:10000,:], Y[:10000], X[10000:20000, :], Y[10000:20000]
```

# Estimators vs Score

```python
np.random.seed(0)
n = [i for i in range(100,1001,100)]
scores = np.zeros(len(n))
for i in range(len(n)):
    learner = RandomForestClassifier(n_estimators=n[i])
    learner.fit(Xtr,Ytr)
    scores[i] = learner.score(Xva,Yva)
plt.plot(n,scores)
plt.show()
```

```python
#x = [i,j for i,j (zip(scores,n))]
print(list(zip(scores,n)))
```

# Max_depth vs score

```python
np.random.seed(0)
d = [i for i in range(1,30)]
dscores = np.zeros(len(d))
for i in range(len(d)):
    learner = RandomForestClassifier(max_depth=d[i])
    learner.fit(Xtr,Ytr)
    dscores[i] = learner.score(Xva,Yva)
plt.plot(d,dscores)
plt.show()
```

```python
np.random.seed(0)
d = [i for i in range(30,61)]
dscores = np.zeros(len(d))
for i in range(len(d)):
    learner = RandomForestClassifier(max_depth=d[i])
    learner.fit(Xtr,Ytr)
    dscores[i] = learner.score(Xva,Yva)
plt.plot(d,dscores)
plt.show()
```

```python
print(list(zip(dscores,d)))
```

```python
learner = RandomForestClassifier(max_depth=24)
learner.fit(Xtr,Ytr)
res = learner.score(Xva,Yva)
print(res)
```
